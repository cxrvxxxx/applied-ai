7. Write a program that checks 1 to N if the number is both divisible by 4 and 6. Print all the numbers that satisfy the condition. If no number (from 1 to N) satisfies the condition, then print "No numbers are divisible by 4 and 6".

Sample Output:


````python
# write your code here
n = int(input("Enter a value for N: "))

d = []

for i in range(1, n):
    if i % 4 == 0 and i % 6 == 0:
        d.append(i)

if d:
    for i in d:
        print(f"{i} is divisible by 4 and 6")
else:
    print("No numbers are divisible by 4 and 6")
````

8. Write a program that finds the sum of the digits of an integer.

Sample Output:


````python
# write your code here
n = input("Enter an integer: ")

sum = 0

for i in n:
    sum += int(i)

print(f"Sum of the digits  = {sum}")
````

9. Write a Python program that outputs the factorial of a number.

Sample Output:


````python
# write your code here 
n = int(input("Enter a number: "))

f = 1

for i in range(1, n + 1):
    f = f * i

print(f"The factorial of {n} is {f}")
````

10. Write a Python program that displays the reverse of the given positive integer. If the integer is negative, output "Invalid Number".

Sample Output:


````python
# write your code here
n = input("Enter a positive integer: ")

print(n[::-1]) if int(n) >= 0 else print("Invalid Number")
````

11. Write a program that checks if the given number is a palindrome. If the number is not a palindrome, then output "The given number is NOT a palindrome number."

Sample Output:


````python
# write your code here
n = input("Enter a number: ")

if n == n[::-1]:
    print("The given number is a palindrome number.")
else:
    print("The given number is NOT a palindrome number.")
````

12. Write a program that checks whether the given integer is a prime number or not.

Sample Outputs:



````python
# write your code here
num = int(input("Enter a number: "))
if num > 1:
    for i in range(2, int(num/2)+1):
        if (num % i) == 0:
            print(num, "is NOT a prime number")
            break
    else:
        print(num, "is a prime number")
else:
    print(num, "is NOT a prime number")
````
